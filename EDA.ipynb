```python
sns.heatmap(df.corr(),annot=True,cmap="viridis")
#There is a Strong correlation between sales and profit
#There is negative correlation between profit and discount
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=df,color="Orange")
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.show()
#plt.tightlayout

#Bookcases,chairs,Tables,Furnishing falls in Furniture category
#labels, Storage, art etc falls in office supplies category
#Phones, Machine ,copiers etc falls in Technology category
```

```python
df.hist(bins=50,figsize=(20,15),color="green")
plt.show()
```

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit',data=df,color='blue')
plt.show()

#When the Dicount is increasing the profit decreses
#Whne the Discount is decreasing the profit is increasing
```

```python
sns.scatterplot(data=df,x='Sales',y='Profit')

#When the sales are increasing profit also incresing
#When the sales are less profit is also less
#In between 8000 to 10000 sales are increasing but profit is decreasing
```

```python
sns.pairplot(df,hue='Sub-Category')
plt.show()
```
